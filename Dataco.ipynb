```python
#Top 10 countries with fraud orders
countryfraudcount = SUSPECTED_FRAUD_Table.groupby('Order Country').size().nlargest(10)
countryfraudcount
```

```python
#Top 10 countries with highest orders count
countryHighcount = MainTable.groupby('Order Country').size().nlargest(10)
countryHighcount
```

```python
# Countries percentages of fraud orders.
data= {'Countriesfraudordercount':SUSPECTED_FRAUD_Table.groupby('Order Country').size(),
       'CountriesHighordercount':MainTable.groupby('Order Country').size(),
      'Percentages':SUSPECTED_FRAUD_Table.groupby('Order Country').size()*100/MainTable.groupby('Order Country').size()}
```

```python
countryFraudDataPercetage=pd.DataFrame(data)
countryFraudDataPercetage
```

```python
countryFraudDataPercetage.mean()
```

```python
#list of cities with higher then mean value orders and fraud percentages 
countryFraudDataPercetage[np.logical_and(countryFraudDataPercetage['CountriesHighordercount']>1100,countryFraudDataPercetage['Percentages']>3)]
```

<b>Findings:</b> <br>
Countries with highest order count have highest fraudulent rate. therefore data is homegenious at country level.
